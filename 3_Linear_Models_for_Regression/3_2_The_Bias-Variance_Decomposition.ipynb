In pratical enviorenment, the training data sets are limited. In this situation,
- If we use complex model, it will lead to severe over-fitting.
- If we use simple model to avoid over-fitting, it will limit the flexibility of the model to capture interesting and important trends in the data.
- If we use regularization, which can control over-fitting for complex model, it will raise the question of how to determine a suitable value for the regularization coefficient $\lambda$.

We shall introduce these model complexity issue in a frequentist viewpoint, known as the *bias-variance* trade-off.

# Loss functions for regression
Consider a linear regression system
- $\mathbf{x}$ is the input of the system.
- $t$ is the corresponding output.
- $p(\mathbf{x},t)$ is a joint distribution that consists of the parameters $\mathbf{x}$ and $t$. 
- $y(\mathbf{x})$ is the function that we seek to suit the system.
- $L(t,y(\mathbf{x}))$ is the loss function for us to judge the goodness of the function $y(\mathbf{x})$. Here we use the squared loss form
$$L(t,y(\mathbf{x}))=\{y(\mathbf{x})-t\}^2$$
- $\mathbb{E}[L]$ is the expected loss, which is given by
$$\begin{align*}
\mathbb{E}[L] &=\iint L(t,y(\mathbf{x}))p(\mathbf{x},t)d\mathbf{x}dt \tag{1.86}\\
&=\iint \{y(\mathbf{x})-t\}^2p(\mathbf{x},t)d\mathbf{x}dt \tag{1.87}
\end{align*}$$

Our goal is to choose $y(\mathbf{x})$ so as to minimize $\mathbb{E}[L]$. From the *calculus of variance*, the functional derivative $\frac{\partial \mathbb{E}[L]}{\partial y}$ should vanish in order to minimize the functional $\mathbb{E}[L]$.
$$\begin{align*}
\frac{\partial \mathbb{E}[L]}{\partial y}&=2\int\{y(\mathbf{x})-t\}p(\mathbf{x},t)dt&=0 \tag{1.88}\\
&=2\int y(\mathbf{x})p(\mathbf{x},t)dt-2\int tp(\mathbf{x},t)dt &=0\\
&=2y(\mathbf{x})p(\mathbf{x})-2\int tp(\mathbf{x},t)dt &=0\\
\Rightarrow &y(\mathbf{x})p(\mathbf{x}) =\int tp(\mathbf{x},t)dt\\
\Rightarrow y(\mathbf{x}) &=\frac{\int tp(\mathbf{x},t)dt}{p(\mathbf{x})} =\int tp(t|\mathbf{x})dt=\mathbb{E}_t[t|\mathbf{x}] \tag{1.89}
\end{align*}$$
where 
- <font color='red'>$\mathbb{E}_t[t|\mathbf{x}]$ represent the mean of the distribution of the target $t$ given the input $\mathbf{x}$.</font> Normally, we can assume that this conditional distribution is a Gaussian which is denoted by $p(t|\mathbf{x})=\mathcal{N}(\mu_{\mathbf{x}}, \beta^{-1})$. 
- <font color='red'>$y(\mathbf{x})=\mathbb{E}_t[t|\mathbf{x}]=\mu_{\mathbf{x}}$</font> means that the function $y(\mathbf{x})$ should be equal to the mean of this conditional distribution so as to minimize the loss.

With the conditional distribution, we can also derive the expected loss in a slightly different way.
$$\begin{align*}
\mathbb{E}[L] &= \iint \{y(\mathbf{x})-t\}^2p(\mathbf{x},t)d\mathbf{x}dt\\
&=\iint \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]+\mathbb{E}[t|\mathbf{x}]-t\}^2p(\mathbf{x},t)d\mathbf{x}dt\\
&=\iint \big\{ \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}^2+2\{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}\{\mathbb{E}[t|\mathbf{x}]-t\}+\{\mathbb{E}[t|\mathbf{x}]-t\}^2\big\}p(\mathbf{x},t)d\mathbf{x}dt\\
&=\underbrace{\iint \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}^2p(\mathbf{x},t)d\mathbf{x}dt}_{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\ is\ not\ related\ to\ t}
+\underbrace{2\iint\{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}\{\mathbb{E}[t|\mathbf{x}]-t\}p(\mathbf{x},t)d\mathbf{x}dt}_{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\ is\ not\ related\ to\ t}
+\iint\{\mathbb{E}[t|\mathbf{x}]-t\}^2p(\mathbf{x},t)d\mathbf{x}dt\\
&=\int \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}^2p(\mathbf{x})d\mathbf{x}
+\int\{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}\left(\int \{\mathbb{E}[t|\mathbf{x}]-t\}p(\mathbf{x},t) dt\right)d\mathbf{x}
+\iint\{\mathbb{E}[t|\mathbf{x}]-t\}^2p(\mathbf{x},t)d\mathbf{x}dt\\
&=\int \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}^2p(\mathbf{x})d\mathbf{x}
+\int\{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}\left(\int \{\mathbb{E}[t|\mathbf{x}]-t\}p(t|\mathbf{x}) dt\right)p(\mathbf{x})d\mathbf{x}
+\iint\{\mathbb{E}[t|\mathbf{x}]-t\}^2p(\mathbf{x},t)d\mathbf{x}dt\\
&=\int \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}^2p(\mathbf{x})d\mathbf{x}
+\int\{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}\big(\mathbb{E}[t|\mathbf{x}]-\mathbb{E}[t|\mathbf{x}]\big)p(\mathbf{x})d\mathbf{x}
+\iint\{\mathbb{E}[t|\mathbf{x}]-t\}^2p(\mathbf{x},t)d\mathbf{x}dt\\
&=\int \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}^2p(\mathbf{x})d\mathbf{x}+0+
\iint\{\mathbb{E}[t|\mathbf{x}]-t\}^2p(\mathbf{x},t)d\mathbf{x}dt \\
&=\color{Orange}{\int \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}^2p(\mathbf{x})d\mathbf{x}}+
\color{Green}{\iint\{\mathbb{E}[t|\mathbf{x}]-t\}^2p(\mathbf{x},t)d\mathbf{x}dt} \tag{3.37}
\end{align*}$$
If we take the assumption that $p(t|\mathbf{x})=\mathcal{N}(\mu_{\mathbf{x}},\beta^{-1})$, then the expected loss is given by
$$\begin{align*}
\mathbb{E}[L]
&=\int \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}^2p(\mathbf{x})d\mathbf{x}+
\iint\{\mathbb{E}[t|\mathbf{x}]-t\}^2p(\mathbf{x},t)d\mathbf{x}dt \\
&=\int \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}^2p(\mathbf{x})d\mathbf{x}+
\iint\{\mathbb{E}[t|\mathbf{x}]-t\}^2p(t|\mathbf{x})p(\mathbf{x})d\mathbf{x}dt\\
&=\underbrace{\int \{y(\mathbf{x})-\mathbb{E}[t|\mathbf{x}]\}^2p(\mathbf{x})d\mathbf{x}}_{y(\mathbf{x})=\mathbb{E}[t|\mathbf{x}]\Rightarrow this\ term\ vanishes}
+\int\left(\int\{\mathbb{E}[t|\mathbf{x}]-t\}^2p(t|\mathbf{x})dt\right) p(\mathbf{x})d\mathbf{x}\\
&=0+\int\beta^{-1} p(\mathbf{x})d\mathbf{x}\\
&=\beta^{-1}
\end{align*}$$


# The Bias-Variance Decomposition
Recall that our goal is to choose $y(\mathbf{x})$ so as to minimize $\mathbb{E}[L]$. 

- $y(\mathbf{x},\mathbf{w})$, our model which is denoted by this function and governed by a parameter vector $\mathbf{w}$.
- $\mathcal{D}$, a training data set. The parameter vector $\mathbf{w}$ is determined by the data set.
- $y(\mathbf{x};\mathcal{D})$, we can rewrite the function $y(\mathbf{x},\mathbf{w})$ to $y(\mathbf{x};\mathcal{D})$. For each data set, we can run our learning algorithm and obtain a prediction function $y(\mathbf{x};\mathcal{D})$. 
- $L$, suppose we have $L$ data sets.
- $N$, suppose each data set has $N$ data points. 

Now we want to evaluate the expected loss from the $L$ data sets.
$$\begin{align*}
\mathbb{E}_{\mathcal{D}}[\mathbb{E}[L]]
&=\mathbb{E}_{\mathcal{D}}\left[\color{Orange}{\int \{y(\mathbf{x};\mathcal{D})-h(\mathbf{x})\}^2p(\mathbf{x})d\mathbf{x}}+
\color{Green}{\iint\{h(\mathbf{x})-t\}^2p(\mathbf{x},t)d\mathbf{x}dt}\right]\quad let\ h(\mathbf{x})=\mathbb{E}[t|\mathbf{x}]\\
&=\mathbb{E}_{\mathcal{D}}\left[\color{Orange}{\int \{y(\mathbf{x};\mathcal{D})-h(\mathbf{x})\}^2p(\mathbf{x})d\mathbf{x}}\right]+
\color{Green}{\iint\{h(\mathbf{x})-t\}^2p(\mathbf{x},t)d\mathbf{x}dt}\\
&=\int \mathbb{E}_{\mathcal{D}}\left[\{y(\mathbf{x};\mathcal{D})-h(\mathbf{x})\}^2\right]p(\mathbf{x})d\mathbf{x}+
\color{Green}{\iint\{h(\mathbf{x})-t\}^2p(\mathbf{x},t)d\mathbf{x}dt}\\
&=\int \mathbb{E}_{\mathcal{D}}\left[\{y(\mathbf{x};\mathcal{D})-\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]+\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]-h(\mathbf{x})\}^2\right]p(\mathbf{x})d\mathbf{x}+
\color{Green}{\iint\{h(\mathbf{x})-t\}^2p(\mathbf{x},t)d\mathbf{x}dt}\\
&=\int \mathbb{E}_{\mathcal{D}}\left[
\{y(\mathbf{x};\mathcal{D})-\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]\}^2
+\{\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]-h(\mathbf{x})\}^2
+2\{y(\mathbf{x};\mathcal{D})-\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]\}\{\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]-h(\mathbf{x})\}
\right]p(\mathbf{x})d\mathbf{x}\\
&\quad +\color{Green}{\iint\{h(\mathbf{x})-t\}^2p(\mathbf{x},t)d\mathbf{x}dt}\\
&=\int 
\mathbb{E}_{\mathcal{D}}\left[\{y(\mathbf{x};\mathcal{D})-\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]\}^2\right]
+\mathbb{E}_{\mathcal{D}}\left[\{\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]-h(\mathbf{x})\}^2\right]\\
&\quad+2\underbrace{\mathbb{E}_{\mathcal{D}}\left[\{y(\mathbf{x};\mathcal{D})-\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]\}\right]}_{=\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]-\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]=0} \{\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]-h(\mathbf{x})\}
p(\mathbf{x})d\mathbf{x}
+\color{Green}{\iint\{h(\mathbf{x})-t\}^2p(\mathbf{x},t)d\mathbf{x}dt}\\
&=\int 
\mathbb{E}_{\mathcal{D}}\left[\{y(\mathbf{x};\mathcal{D})-\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]\}^2\right]
+\{\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]-h(\mathbf{x})\}^2
p(\mathbf{x})d\mathbf{x}
+\color{Green}{\iint\{h(\mathbf{x})-t\}^2p(\mathbf{x},t)d\mathbf{x}dt}\\
&=
\underbrace{\color{Blue}{\int \{\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]-h(\mathbf{x})\}^2p(\mathbf{x})d\mathbf{x}}}_{\color{Blue}{(bias)^2}}
+\underbrace{\color{Red}{\int \mathbb{E}_{\mathcal{D}}\left[\{y(\mathbf{x};\mathcal{D})-\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]\}^2\right]p(\mathbf{x})d\mathbf{x}}}_{\color{Red}{variance}}
+\underbrace{\color{Green}{\iint\{h(\mathbf{x})-t\}^2p(\mathbf{x},t)d\mathbf{x}dt}}_{\color{Green}{noise}}\tag{3.41}
\end{align*}$$
We can see that the expected loss is the sum of the $\color{Blue}{(bias)^2}$ and $\color{Red}{variance}$ and $\color{Green}{noise}$, where
- $\color{Blue}{(bias)^2}$, called the squared bias, represents the extent to which the average prediction over all data sets differs from the desired regression function.
- $\color{Red}{variance}$ measures the extent to which the solutions for individual data sets vary around there average, and hence this measures the extent to which the function $y(\mathbf{x};\mathcal{D})$ is sensitive to the particular choice of data set.
- $\color{Green}{noise}$ is the variance of the intrinsic noise on generating the data. This noise also represents the minimun achivable value of the expected loss.


# Single-variant Example

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
#from mpl_toolkits.mplot3d import Axes3D
#fig = plt.figure()
#ax = fig.gca(projection='3d')

D = 100
N = 25
M = 25
L = 20

# proper designed basis functions will help us achive curve fitting easily
S = 0.2
MU = np.linspace(-1.5, 1.5, M-1)
Gaussian_Basis = lambda x: np.exp(-(x-MU)**2/(2*S**2))

def BuildPhi(x):
    Phi = np.array([])
    for k in range(len(x)):
        Phi_line = Gaussian_Basis(x[k])
        #print(Phi_line)
        if len(Phi) == 0:
            Phi = np.copy(Phi_line)
        else:
            Phi = np.vstack((Phi, Phi_line))
    Phi = np.c_[np.ones((len(x), 1)), Phi]
    return Phi

def draw_one_graph(ax, Phi, lamb):
    x_data = np.linspace(0,1,N)
    #for drawing
    x = np.linspace(0,1,D)
    y = np.zeros(x.shape)
    
    S = np.array([])
    
    for l in range(L):
        noise = np.random.normal(loc=0,scale=0.5, size=x_data.shape)
        t_data = np.sin(2*np.pi*x_data) + noise
        #ax.scatter(x_data, t_data, s=100, facecolors='none', edgecolors='blue', label='training data')
        weight = np.linalg.inv(lamb * np.identity(M) + Phi.T @ Phi) @ Phi.T @ t_data.T
        #print(weight)
        for k in range(D):
            basis = Gaussian_Basis(x[k])
            basis = np.hstack(([1], basis))
            y[k] = np.sum(weight * basis)
        line_y, = ax.plot(x,y, color='C3', alpha=0.5)
        if len(S) == 0:
            S = np.copy(y)
        else:
            S += y
    ave_y = S/L
    line_e, = ax.plot(x, ave_y, color='k')
    ax.legend([line_y, line_e], [r'$y(\mathbf{x};\mathcal{D})$', r'$\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]$'])
    return x, ave_y

def testPhi(Phi):
    np.set_printoptions(formatter={'float': '{: 0.3f}'.format})
    print(Phi)
    a = Phi
    B = np.linalg.pinv(a)
    print(B)
    print(np.allclose(a, np.dot(a, np.dot(B, a))))
    print(np.linalg.cond(Phi))
    print(np.linalg.det(Phi))

    
def main():
    fig = plt.figure(figsize=(15,15), dpi=60)
    ax1 = fig.add_subplot(3,2,1)
    ax2 = fig.add_subplot(3,2,2)
    ax3 = fig.add_subplot(3,2,3)
    ax4 = fig.add_subplot(3,2,4)
    ax5 = fig.add_subplot(3,2,5)
    ax6 = fig.add_subplot(3,2,6)
    
    ax1.set_ylim(-1.5, 1.5)
    ax3.set_ylim(-1.5, 1.5)
    ax5.set_ylim(-1.5, 1.5)
    x_data = np.linspace(0,1,N)
    Phi = BuildPhi(x_data)
    
    #testPhi(Phi)
    
    x = np.linspace(0,1,D)
    x_r, y_r = draw_one_graph(ax1, Phi, 5)
    ax1.set_xlabel("variance")
    ax1.set_ylabel(r"$\lambda=5$")
    ax2.set_xlabel("bias")
    ax2.plot(x, np.sin(2*np.pi*x), label=r'$h(x)$')
    ax2.plot(x_r, y_r, color='k', label=r'$\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]$')
    ax2.legend()
    
    x_r, y_r = draw_one_graph(ax3, Phi, 1)
    ax3.set_xlabel("variance")
    ax3.set_ylabel(r"$\lambda=1$")
    ax4.set_xlabel("bias")
    ax4.plot(x, np.sin(2*np.pi*x),  label=r'$h(x)$')
    ax4.plot(x_r, y_r, color='k', label=r'$\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]$')
    ax4.legend()
    
    x_r, y_r = draw_one_graph(ax5, Phi, 0.001)
    ax5.set_xlabel("variance")
    ax5.set_ylabel(r"$\lambda=0.001$")
    ax6.set_xlabel("bias")
    ax6.plot(x, np.sin(2*np.pi*x),  label=r'$h(x)$')
    ax6.plot(x_r, y_r, color='k', label=r'$\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]$')
    ax6.legend()
    
    plt.show()
    
if __name__ == "__main__":
    main()
```

The right hand-side average curve, which is denoted by $\bar{y}$ or $\mathbb{E}_{\mathcal{D}}[y]$, in the situation of fixed data sets, will be determined by $\lambda$ (the parameter of the regularization), $\mu$ and $\sigma$ (the parameters of the basis functions). So a well defined basis function as well as regularization parameter will be significant.


# Calculation of each component of Loss
We can also use this single-variant example to calculate each component of the loss.  
The average prediction is estimated from
$$\mathbb{E}_{\mathcal{D}}[y(\mathbf{x};\mathcal{D})]=\bar{y}(x)=\frac{1}{L}\sum_{l=1}^Ly^{(l)}(x) \tag{3.45}$$
and the integrated squared bias and integrated variance are then given by
$$\begin{align*}
(bias)^2 &= \frac{1}{N}\sum_{n=1}^N\{\bar{y}(x_n)-h(x_n)^2\} \tag{3.46} \\
variance &= \frac{1}{N}\sum_{n=1}^N\frac{1}{L}\sum_{l=1}^L\{y^{(l)}(x_n)-\bar{y}(x_n)\} \tag{3.47}
\end{align*}$$


```python
def gen_b_and_v(x_data, y_data, T, Phi, lamb):
    y = np.zeros(x_data.shape)
    S = np.zeros(x_data.shape)
    V = np.zeros(x_data.shape)
    Y = np.array([])
    
    for t_data in T:
        weight = np.linalg.inv(lamb * np.identity(M) + Phi.T @ Phi) @ Phi.T @ t_data.T
        for n in range(N):
            basis = Gaussian_Basis(x_data[n])
            basis = np.hstack(([1], basis))
            y[n] = np.sum(weight * basis)
        
        if len(Y) == 0:
            Y = np.copy(y)
        else:
            Y = np.vstack((Y, y))
        S += y
        
    avg_y = S/L
    for yl in Y:
        V += (yl-avg_y)**2
            
    b = np.average((y_data - avg_y)**2)
    v = np.average(V/L)
    return v, b

def main():
    fig = plt.figure(figsize=(6,4), dpi=60)
    ax = fig.add_subplot(1,1,1)
    
    x_data = np.linspace(0,1,N)
    Phi = BuildPhi(x_data)
    
    bias = np.array([])
    variance = np.array([])
    
    lnlambs = np.linspace(-5, 5, 100)
    
    x_data = np.linspace(0,1,N)
    y_data = np.sin(2*np.pi*x_data)
    T = np.array([])
    for l in range(L):
        noise = np.random.normal(loc=0,scale=0.5, size=x_data.shape)
        t_data = y_data + noise
        if len(T) == 0:
            T = np.copy(t_data)
        else:
            T = np.vstack((T, t_data))

    for lnlamb in lnlambs:
        lamb = np.exp(lnlamb)
        v, b = gen_b_and_v(x_data, y_data, T, Phi, lamb)
        bias = np.hstack((bias, b))
        variance = np.hstack((variance, v))
    ax.plot(lnlambs, bias, label=r"$(bias)^2$")
    ax.plot(lnlambs, variance, label=r"$variance$")
    ax.plot(lnlambs, bias+variance, label=r"$(bias)^2+variance$")
    ax.set_xlabel(r"$\ln\lambda$")
    ax.legend()

if __name__ == "__main__":
    main()
```

# Conclusion
1. High regularization coefficient $\lambda$ (less complex model) gives low variance (because the red curves in the left plot look similar) but high bias (because the two curves in the right plot are very different).
2. Low regularization coefficient $\lambda$ (more complex model) gives high variance but low bias.


# Limitation
Although the bias-variance decomposiion **provide us the interesting insights into the model complexity issue** from a frequentist perspective, it is of limited practical value, because it is based on averages with respect to ensembles of data sets, whereas in pratical we hav only the single observed data set. <font color='red'>If we had a large number of independent training sets of given size, we would be better off combining them into a single large training set, which of course would reduce the level of over-fitting for a given model complexity.</font>
