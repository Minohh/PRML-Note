# Marginal Gaussian distributions

We can use the same approach that we have done in the conditional Gaussian distributions. That is converting the marginal Gaussian distribution exponent to the general Gaussian distribution exponent to get the mean and covariance.

The marginal Gaussian distribution is given by
$$p(\mathbf{x}_a)=\int p(\mathbf{x}_a,\mathbf{x}_b)d\mathbf{x}_b$$
which, as we shall see, is also a Gaussian.

### General distribution exponent
$$\begin{align*}-\frac{1}{2}(\mathbf{x}-\mathbf{\mu})^T\Sigma^{-1}(\mathbf{x}-\mathbf{\mu})
&=-\frac{1}{2}\mathbf{x}^T\Sigma^{-1}\mathbf{x}+\mathbf{x}^T\Sigma^{-1}\mathbf{\mu}-\frac{1}{2}\mathbf{\mu}^T\Sigma^{-1}\mathbf{\mu}
\end{align*}$$
In this equation, 
- $-\frac{1}{2}\mathbf{x}^T\Sigma^{-1}\mathbf{x}$ is the second order term.
- $\mathbf{x}^T\Sigma^{-1}\mathbf{\mu}$ is the first order term.
- $-\frac{1}{2}\mathbf{\mu}^T\Sigma^{-1}\mathbf{\mu}$ is the constant term.  

### Marginal distribution exponent
The marginal distribution $p(\mathbf{x}_a)$ can be denoted by
$$p(\mathbf{x}_a)=\mathcal{N}(\mathbf{x}|\mathbf{\mu}_a,\Sigma_a)=\int p(\mathbf{x}_a,\mathbf{x}_b)d\mathbf{x}_b=\int\frac{1}{(2\pi)^{D/2}}\frac{1}{|\Sigma|^{1/2}}exp\left\{-\frac{1}{2}(\mathbf{x}-\mathbf{\mu}^T)\Sigma^{-1}(\mathbf{x}-\mathbf{\mu})\right\}d\mathbf{x}_b$$
where the exponent is a quadratic form
$$\begin{align*}
-\frac{1}{2}(\mathbf{x}-\mathbf{\mu})^T\Sigma^{-1}(\mathbf{x}-\mathbf{\mu})
&=-\frac{1}{2}
\left[\left( \begin{matrix}\mathbf{x}_a\\ \mathbf{x}_b\end{matrix}\right )
-\left( \begin{matrix}\mathbf{\mu}_a\\ \mathbf{\mu}_b\end{matrix}\right )\right]^T
\left( \begin{matrix}\Lambda_{aa} &\Lambda_{ab} \\ \Lambda_{ba} & \Lambda_{bb}\end{matrix}\right )
\left[\left( \begin{matrix}\mathbf{x}_a\\ \mathbf{x}_b\end{matrix}\right )
-\left( \begin{matrix}\mathbf{\mu}_a\\ \mathbf{\mu}_b\end{matrix}\right )\right]\\
&=-\frac{1}{2}(\mathbf{x}_a-\mathbf{\mu}_a)^T\Lambda_{aa}(\mathbf{x}_a-\mathbf{\mu}_a)
-\frac{1}{2}(\mathbf{x}_a-\mathbf{\mu}_a)^T\Lambda_{ab}(\mathbf{x}_b-\mathbf{\mu}_b)\\
&\quad -\frac{1}{2}(\mathbf{x}_b-\mathbf{\mu}_b)^T\Lambda_{ba}(\mathbf{x}_a-\mathbf{\mu}_a)
-\frac{1}{2}(\mathbf{x}_b-\mathbf{\mu}_b)^T\Lambda_{bb}(\mathbf{x}_b-\mathbf{\mu}_b)\\
\end{align*}
$$

#### Extract the $\mathbf{x}_b$ terms
Extract the terms that involve $\mathbf{x}_b$ from this exponent, we have
$$-\frac{1}{2}\mathbf{x}_b^T\Lambda_{bb}\mathbf{x}_b+\mathbf{x}_b^T\mathbf{m}
=-\frac{1}{2}(\mathbf{x}_b-\Lambda_{bb}^{-1}\mathbf{m})^T\Lambda_{bb}(\mathbf{x}_b-\Lambda_{bb}^{-1}\mathbf{m})+\frac{1}{2}\mathbf{m}^T\Lambda_{bb}^{-1}\mathbf{m}\qquad \mathbf{m}=\Lambda_{bb}\mathbf{\mu}_b-\Lambda_{ba}(\mathbf{x}_a)-\mathbf{\mu}_a$$
We see that the dependence on $\mathbf{x}_b$ has been cast into the standard quadratic form of a general Gaussian distribution. Thus, when we take the exponential of this quadratic form, we see that the integration over $\mathbf{x}_b$
$$\int exp\left\{-\frac{1}{2}(\mathbf{x}_b-\Lambda_{bb}^{-1}\mathbf{m})^T\Lambda_{bb}(\mathbf{x}_b-\Lambda_{bb}^{-1}\mathbf{m}) \right\}d\mathbf{x}_b$$
This term will be integrated to be a constant and as a result, will be the reciprocal of the nomalization coefficient.  

#### The remaining $\mathbf{x}_a$ terms
The remaining exponent that depend on $\mathbf{x}_a$ is given by
$$\begin{align*}
&\quad\frac{1}{2}\big[ \Lambda_{bb}\mathbf{\mu}_b -\Lambda_{ba}(\mathbf{x}_a-
\mathbf{\mu}_a)\big]^T\Lambda_{bb}^{-1}\big[\Lambda_{bb}\mathbf{\mu}_b-\Lambda_{ba}(\mathbf{x}_a-\mathbf{\mu}_a)\big]
-\frac{1}{2}\mathbf{x}_a^{T}\Lambda_{aa}\mathbf{x}_a+\mathbf{x}^T(\Lambda_{aa}\mathbf{\mu}_a+\Lambda_{ab}\mathbf{\mu}_b)+const\\
&=-\frac{1}{2}\mathbf{x}_a^T(\Lambda_{aa}-\Lambda_{ab}\Lambda_{bb}^{-1}\Lambda_{ba})\mathbf{x}_a
+\mathbf{x}_a^T(\Lambda_{aa}\Lambda_{bb}^{-1}\Lambda_{ba})^{-1}\mathbf{\mu}_a+const\end{align*}$$
Thus, the covariance of the marginal distribution of $p(\mathbf{x}_a)$ is given by
$$\Sigma_a=(\Lambda_{aa}-\Lambda_{ab}\Lambda_{bb}^{-1}\Lambda_{ba})^{-1}=\Sigma_{aa}$$
where the second equation can be concluded from the last article.  
Similarly, the mean is given by
$$\mathbf{\mu}_{a}=\Sigma_a(\Lambda_{aa}-\Lambda_{ab}\Lambda_{bb}^{-1}\Lambda_{ba})\mathbf{\mu}_a=\mathbf{\mu}_a$$

#### Conclusion
<font color='red'>$$\begin{align*}
\mathbb{E}[\mathbf{x}_a] &=\mu_{a}\\
cov[\mathbf{x}_a]&=\Sigma_{aa}\\
p(\mathbf{x}_a)&=\mathcal{N}(\mathbf{x}_a|\mathbf{\mu}_a,\Sigma_{aa})
\end{align*}$$</font>
