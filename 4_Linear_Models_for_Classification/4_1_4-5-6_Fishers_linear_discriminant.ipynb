# Definition

## Dimensionality reduction

$$y = \mathbf{w}^T \mathbf{x}\tag{4.20}$$
where the vector $\mathbf{x}$ is projected to the vector $\mathbf{w}$, and the result is $y$.

For the case of two classes, we can determine a threshold on $y$ and classify $y\geqslant -w_0$ as class $C_1$, and otherwise class $C_2$.

In this article, our goal is to determine the vector $\mathbf{w}$ and the threshold $w_0$.


## Determine $\mathbf{w}$

Poor selected $\mathbf{w}$ will lead to strongly overlapping on projection, whereas well selected $\mathbf{w}$ can select a projection that maximizes the class seperation.
### Simplest approach
The simplest measure of the seperation of the class, when projected onto $\mathbf{w}$, is the separation of the projected class means. This suggests that we might choose $\mathbf{w}$ so as to maximize
<font color='Red'>$$m_2-m_1 = \mathbf{w}^T(\mathbf{m}_2-\mathbf{m}_1) \tag{4.22}$$</font>
where $\mathbf{m}_1$ and $\mathbf{m}_2$ are the mean vectors of the points of class $C_1$ and $C_2$.
$$\mathbf{m}_1 = \frac{1}{N_1}\sum_{n\in C_1}\mathbf{x}_n,\qquad \mathbf{m}_2=\frac{1}{N_2}\sum_{n\in C_2}\mathbf{x}_n \tag{4.21}$$
and $m_k$ in the projected point of class $k$
$$m_k = \mathbf{w}^T\mathbf{m}_k \tag{4.23}$$

However, (4.22) can be made arbitarily large simply by increasing the magnitude of $\mathbf{w}$. To solve this problem, we could constrain $\mathbf{w}$ to have unit length, so that $\mathbf{w}^T\mathbf{w}=1$. We can use the largrange multiplier to perform the constrained maximization.

$$\begin{align*}
J(\mathbf{w}) &= \mathbf{w}^T(\mathbf{m}_2-\mathbf{m}_1) + \lambda \mathbf{w}^T\mathbf{w}\\
\frac{dJ(\mathbf{w})}{\mathbf{w}} &= (\mathbf{m}_2-\mathbf{m}_1)+\lambda\mathbf{w} = 0\\
\Rightarrow \color{red}{\mathbf{w}} &\color{Red}{\propto (\mathbf{m}_2-\mathbf{m}_1)}
\end{align*}$$
which infers that $\mathbf{w}$ and $\mathbf{m}_2-\mathbf{m}_1$ have the same direction.

However, this approach still has a **problem**. That is the considerable overlap when projected, which is arises from the strongly nondiagonal covariances of the class distribution.

### Fisher criterion

There is an idea proposed by Fisher to solve the overlap problem above, that is to take the variance of the projected data into account. <font color='Red'>The best $\mathbf{w}$ will give a large separation between the projected class means and also gives a small variance within each class.</font> Thus the Fisher criterion that we need to maximize is given by
$$J(\mathbf{w}) = \frac{(m_2-m_1)^2}{s_1^2+s_2^2} \tag{4.25}$$
where $s_k^2$ is the variance of the projected data of class $C_k$.
$$s_k^2 = \sum_{n\in C_k}(y_n-m_k)^2 \tag{4.24}$$

We can make the dependence on $\mathbf{w}$ explicit by using (4.20), (4.23), and (4.24) t orewrite the Fisher criterion in the form

<font color='Red'>$$J(\mathbf{w})=\frac{\mathbf{w}^TS_B\mathbf{w}}{\mathbf{w}^TS_W\mathbf{w}}\tag{4.26}$$</font>
where $S_B$ is the *between-class* covariance matrix and is given by
$$S_B = (\mathbf{m}_2-\mathbf{m}_1)(\mathbf{m}_2-\mathbf{m}_1)^T \tag{4.27}$$
and $S_W$ is the total *within-class* covariance matrix, given by
$$S_W = \sum_{n\in C_1}(\mathbf{x}_n-\mathbf{m}_1)(\mathbf{x}_n-\mathbf{m}_1)^T+\sum_{n\in C_2}(\mathbf{x}_n-\mathbf{m}_2)(\mathbf{x}_n-\mathbf{m}_2)^T \tag{4.28}$$

Differentiating (4.26) with respect to $\mathbf{w}$, we find that $J(\mathbf{w})$ is maximized when

$$\begin{align*}J(\mathbf{w})\mathbf{w}^TS_W\mathbf{w} &= \mathbf{w}^TS_B\mathbf{w}\\
\frac{d}{d\mathbf{w}} J(\mathbf{w})\mathbf{w}^TS_W\mathbf{w} &= \frac{d}{d\mathbf{w}}\mathbf{w}^TS_B\mathbf{w}\\
J'(\mathbf{w})\mathbf{w}^TS_W\mathbf{w}+J(\mathbf{w})S_W\mathbf{w}+J(\mathbf{w})S_W\mathbf{w} &= S_B\mathbf{w}+S_B\mathbf{w}\\
\underbrace{J'(\mathbf{w})\mathbf{w}^TS_W\mathbf{w}}_{J'(\mathbf{w})=0\ to\ maximize\ J(\mathbf{w})}+\frac{\mathbf{w}^TS_B\mathbf{w}}{\mathbf{w}^TS_W\mathbf{w}}S_W\mathbf{w}+\frac{\mathbf{w}^TS_B\mathbf{w}}{\mathbf{w}^TS_W\mathbf{w}}S_W\mathbf{w} &= S_B\mathbf{w}+S_B\mathbf{w}\\
(\mathbf{w}^TS_B\mathbf{w})S_W\mathbf{w} &= (\mathbf{w}^TS_W\mathbf{w})S_B\mathbf{w} \tag{4.29}
\end{align*}$$

From (4.27), we see that $S_B\mathbf{w}$ is always in the direction of $(\mathbf{m}_2-\mathbf{m}_1)$ because the result of $(\mathbf{m}_2-\mathbf{m}_1)^T\mathbf{w}$ is a scalar. Furthermore, we do not care about the magnitude of $\mathbf{w}$, only its direction, and so tw can drop the scalar factors $(\mathbf{w}^TS_B\mathbf{w})$ and $(\mathbf{w}^TS_W\mathbf{w})$. Multiplying both sides of (4.29) by $S_W^{-1}$ we obtain
<font color='Red'>$$\mathbf{w}\propto S_W^{-1}(\mathbf{m}_2-\mathbf{m}_1) \tag{4.30}$$</font>

which is called *Fisher's linear discriminant*.

## Determine $w_0$
With the vector $\mathbf{w}$, we can obtain the projected value $y$ from arbitary $\mathbf{x}$. Here, we want a threshold $w_0$ to be a boundary in order to classify the projected data $y$.

There is a typical way to find the value of $w_0$. We can model the class-conditional densities $p(y|C_k)$ using Gaussian distributions and then use the techniques of Section 1.2.4 to find the mean and the variance of the Gaussian distribution.
$$\begin{align*}
\mu_{ML}&=\frac{1}{N}\sum_{n=1}^{N}x_n \tag{1.55}\\
\sigma^2_{ML}&=\frac{1}{N}\sum_{n=1}^N(x_n-\mu_{ML})^2 \tag{1.56}
\end{align*}$$
Having found the approximations of the Gaussian distributions, the formalism of Section 1.5.1 then gives an expression for the optimal threshold, which can be briefly conclude as finding the cross point of these two Gaussians.

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.transforms import Affine2D
import mpl_toolkits.axisartist.floating_axes as floating_axes

limX = [-3.0, 5.0]
limY = [-3.0, 5.0]

S1 = np.array([[1.0,.55],
               [.55,0.7]])
S2 = np.array([[.4,.4],
               [.4,.7]])

M1 = np.array([1, 3])
M2 = np.array([3, 2])

def gen_data(n):
    X1 = np.random.multivariate_normal(mean=M1, cov=S1, size=n)
    X2 = np.random.multivariate_normal(mean=M2, cov=S2, size=n)
    return X1, X2

def plot_line(ax, w, Y1, Y2):
    x = np.linspace(limX[0], limX[1], 10)
    #1.24 mean and variance for Gaussian distribution
    my1 = np.average(Y1)
    sy1 = np.average((Y1 - my1)**2)
    my2 = np.average(Y2)
    sy2 = np.average((Y2 - my2)**2)
    Y = np.linspace(min(my1, my2), max(my1, my2), 100)
    
    #1.51 the cross point between these 2 distributions is the decision surface
    dist_y1 = 1.0/(2*np.pi*sy1)**0.5 * np.exp(-0.5/(sy1**2) * (Y-my1)**2)
    dist_y2 = 1.0/(2*np.pi*sy2)**0.5 * np.exp(-0.5/(sy2**2) * (Y-my2)**2)
    index = np.argmin(np.abs(dist_y1 - dist_y2))
    y0 = Y[index]

    y = -(w[0] * x) / w[1] - y0
    ax.plot(x, y, color='green')

def plot_points(ax, X1, X2):
    ax.scatter(X1.T[0], X1.T[1], s=10, color='red')
    ax.scatter(X2.T[0], X2.T[1], s=10, color='blue')
    
    m1 = np.average(X1.T, axis=1)
    m2 = np.average(X2.T, axis=1)
    ax.scatter(m1[0], m1[1], s=80, marker='x', color='red')
    ax.scatter(m2[0], m2[1], s=80, marker='+', color='blue')
    return m1, m2

def setup_axes(ax, pos, w, ymax, ymin):
    fig = ax.get_figure()

    tr = Affine2D().scale(10, 1).rotate(np.arctan(w[1]/w[0]))

    grid_helper = floating_axes.GridHelperCurveLinear(
        tr, extremes=(ymin, ymax, 0, 20))
        
    hist_ax = floating_axes.FloatingSubplot(fig, pos, grid_helper=grid_helper)
    hist_ax.set_frame_on(False)
    hist_ax.axis["bottom"].set_visible(False)
    hist_ax.axis["top"].set_visible(False)
    hist_ax.axis["left"].set_visible(False)
    hist_ax.axis["right"].set_visible(False)

    fig.add_subplot(hist_ax)
    aux_ax = hist_ax.get_aux_axes(tr)

    return hist_ax, aux_ax

def draw_bars(ax, pos, w, Y1, Y2):
    Y = np.array((Y1, Y2))
    ymax = np.amax(Y)
    ymin = np.amin(Y)
    ymax = ymax + (ymax - ymin)/8
    ymin = ymin - (ymax - ymin)/8
    ax_label, ax_plot = setup_axes(ax, pos, w, ymax, ymin)
    
    b = np.linspace(ymin, ymax, 20)
    
    ax_plot.hist(Y1, bins=b, color='red', histtype='barstacked', rwidth=0.8, edgecolor="k")
    ax_plot.hist(Y2, bins=b, color='blue', histtype='barstacked', rwidth=0.8, edgecolor="k")
    
def classify_with_mean(ax, X1, X2):
    m1, m2 = plot_points(ax, X1, X2)
    w = m2-m1
    w = w/np.sqrt(w @ w.T)
    Y1 = w @ X1.T
    Y2 = w @ X2.T
    draw_bars(ax, 121, w, Y1, Y2)
    plot_line(ax, w, Y1, Y2)
    return
    
def classify_with_mean_and_variance(ax, X1, X2):
    m1, m2 = plot_points(ax, X1, X2)
    SW = (X1-m1).T @ (X1-m1) + (X2-m2).T @ (X2-m2)
    w = np.linalg.inv(SW) @ (m2-m1).T
    w = w/np.sqrt(w @ w.T)
    Y1 = w @ X1.T
    Y2 = w @ X2.T
    draw_bars(ax, 122, w, Y1, Y2)
    plot_line(ax, w, Y1, Y2) 
    return

def main():
    fig = plt.figure(figsize=(15,6), dpi=60)
    ax1 = fig.add_subplot(1,2,1)
    ax1.set_xlim(limX[0], limX[1])
    ax1.set_ylim(limY[0], limY[1])
    
    ax2 = fig.add_subplot(1,2,2)
    ax2.set_xlim(limX[0], limX[1])
    ax2.set_ylim(limY[0], limY[1])
    
    X1, X2 = gen_data(50)
    classify_with_mean(ax1, X1, X2)
    classify_with_mean_and_variance(ax2, X1, X2)
    
    
    plt.show()


if __name__ == "__main__":
    main()
```

# Relation to least squares

The sum-of-squares error function can be written
$$E = \frac{1}{2}\sum_{n=1}^N(\mathbf{w}^T\mathbf{x}_n+w_0-t_n)^2 \tag{4.31}$$

## Evaluation of $w_0$
Setting the derivatives of $E$ with respect to $w_0$ to zero, we obtain
$$\begin{align*}
\frac{dE}{dw_0} &= \sum_{n=1}^N(\mathbf{w}^T\mathbf{x}_n+w_0-t_n) = 0 \tag{4.32}\\
&=\sum_{n=1}^N(\mathbf{w}^T\mathbf{x}_n+w_0)-N_1\frac{N}{N_1}+N_2\frac{N}{N_2} \qquad let\ t_n =\left\{ 
\begin{array}{ll}N/N_1, &if\ \mathbf{x}_n\in C_1\\-N/N_2, &if\ \mathbf{x}_n\in C_2\end{array}\right.\\
&=\mathbf{w}^T \sum_{n=1}^N\mathbf{x}_n+Nw_0\\
&=\mathbf{w}^TN\mathbf{m}+Nw_0\\
&\Rightarrow \color{Red}{w_0} \color{Red}{=-\mathbf{w}^T\mathbf{m}} \tag{4.34}
\end{align*}$$

## Evaluation of $\mathbf{w}$
Setting the derivatives of $E$ with respect to $\mathbf{w}$ to zero, we obtain
$$\begin{align*}
\frac{dE}{d\mathbf{w}} &= \sum_{n=1}^N(\mathbf{w}^T\mathbf{x}_n+w_0-t_n)\mathbf{x}_n = 0 \tag{4.33}\\
&=\sum_{n=1}^N(\mathbf{w}^T\mathbf{x}_n-\mathbf{w}^T\mathbf{m})\mathbf{x}_n-\frac{N}{N_1}\sum_{n\in C_1}\mathbf{x}_n+\frac{N}{N_2}\sum_{n\in C_2}\mathbf{x}_n \qquad let\ t_n =\left\{ 
\begin{array}{ll}N/N_1, &if\ \mathbf{x}_n\in C_1\\-N/N_2, &if\ \mathbf{x}_n\in C_2\end{array}\right.\\
&=\sum_{n=1}^N\mathbf{w}^T(\mathbf{x}_n-\mathbf{m})\mathbf{x}_n-N(\mathbf{m}_1-\mathbf{m}_2)\\
&=\sum_{n=1}^N \mathbf{x}_n\mathbf{w}^T(\mathbf{x}_n-\mathbf{m})-N(\mathbf{m}_1-\mathbf{m}_2)\\
&=\sum_{n=1}^N \mathbf{x}_n(\mathbf{x}_n-\mathbf{m})^T\mathbf{w}-N(\mathbf{m}_1-\mathbf{m}_2)\\
&=\left\{\sum_{n=1}^N \Big(\mathbf{x}_n\mathbf{x}_n^T-\mathbf{x}_n\mathbf{m}^T\Big)\right\}\mathbf{w}-N(\mathbf{m}_1-\mathbf{m}_2)\\
&=\left\{\sum_{n=1}^N \Big(\mathbf{x}_n\mathbf{x}_n^T\Big)-N\mathbf{m}\mathbf{m}^T\right\}\mathbf{w}-N(\mathbf{m}_1-\mathbf{m}_2)\\
&=\left\{\left(\sum_{n\in C_1}\mathbf{x}_n\mathbf{x}_n^T-N_1\mathbf{m}_1\mathbf{m}^T\right)
+\left(\sum_{n\in C_2}\mathbf{x}_n\mathbf{x}_n^T-N_2\mathbf{m}_2\mathbf{m}^T\right)\right\}\mathbf{w}-N(\mathbf{m}_1-\mathbf{m}_2)\\
&=\left\{\left(\sum_{n\in C_1}\mathbf{x}_n\mathbf{x}_n^T-N_1\mathbf{m}_1\mathbf{m}_1^T\right)
+\left(\sum_{n\in C_2}\mathbf{x}_n\mathbf{x}_n^T-N_2\mathbf{m}_2\mathbf{m}_2^T\right)\right\}\mathbf{w}-N(\mathbf{m}_1-\mathbf{m}_2)\\
&\quad +\Big\{-N_1\mathbf{m}_1\mathbf{m}^T+N_1\mathbf{m}_1\mathbf{m}_1^T-N_2\mathbf{m}_2\mathbf{m}^T+N_2\mathbf{m}_2\mathbf{m}_2^T\Big\}\mathbf{w}\\
&=S_W\mathbf{w}-N(\mathbf{m}_1-\mathbf{m}_2)\\
&\quad +\Big\{-N_1\mathbf{m}_1\frac{N_1\mathbf{m}_1^T+N_2\mathbf{m}_2^T}{N}+N_1\mathbf{m}_1\mathbf{m}_1^T-N_2\mathbf{m}_2\frac{N_1\mathbf{m}_1^T+N_2\mathbf{m}_2^T}{N}+N_2\mathbf{m}_2\mathbf{m}_2^T\Big\}\mathbf{w}\\
&=S_W\mathbf{w}-N(\mathbf{m}_1-\mathbf{m}_2)\\
&\quad +\Big\{\frac{-N_1N_2\mathbf{m}_1\mathbf{m}_2^T
-N_1N_2\mathbf{m}_2\mathbf{m}_1^T+N_1N_2\mathbf{m}_1\mathbf{m}_1^T+N_1N_2\mathbf{m}_2\mathbf{m}_2^T}{N}\Big\}\mathbf{w}\\
&=S_W\mathbf{w}-N(\mathbf{m}_1-\mathbf{m}_2)+\Big\{\frac{N_1N_2(\mathbf{m}_2-\mathbf{m}_1)(\mathbf{m}_2-\mathbf{m}_1)^T}{N}\Big\}\mathbf{w}\\
&=\left(S_W+\frac{N_1N_2}{N}S_B\right)\mathbf{w}-N(\mathbf{m}_1-\mathbf{m}_2)\\
&\Rightarrow \color{Red}{\left(S_W+\frac{N_1N_2}{N}S_B\right)\mathbf{w}= N(\mathbf{m}_1-\mathbf{m}_2)} \tag{4.37}
\end{align*}$$
Using (4.27), we note that $S_B\mathbf{w}$ is always in the direction of $(\mathbf{m}_2-\mathbf{m}_1)$. Thus we can write
<font color='Red'>$$\mathbf{w}\propto S_W^{-1}(\mathbf{m}_2-\mathbf{m}_1) \tag{4.38}$$</font>
which is the *Fisher's linear discriminant* we discussed before.

## Conclusion
Then it can be concluded that, with the definition 
$$t_n =\left\{ 
\begin{array}{ll}N/N_1, &if\ \mathbf{x}_n\in C_1\\-N/N_2, &if\ \mathbf{x}_n\in C_2\end{array}\right.$$
the sum-of-squares gives the same $\mathbf{w}$ as the Fisher's linear discriminant, and also gives the value of $w_0$ from the equation
$$w_0 =-\mathbf{w}^T\mathbf{m}$$

# Multiple classes
Consider a $K$ classes classifier.

$$\mathbf{y} = \mathbf{W}^T \mathbf{x} \tag{4.39}$$

where 

- $\mathbf{x}$ is an input point.
- $\mathbf{W}$ is a matrix for which columns are the weight vectors $\mathbf{w}_k$, where $k=1,\cdots,K$.
- $\mathbf{y}$ is a vector, and the element $y_k$ in it denotes the projection from a point $\mathbf{x}$ to the weight vector $\mathbf{w}_k$.

Then we can rewrite expression of (4.39) to the form 
$$\begin{bmatrix}y_1\\ y_2\\ \vdots\\ y_{K}\end{bmatrix}
= \begin{bmatrix}\mathbf{w}_1\\ \mathbf{w}_2\\ \vdots\\ \mathbf{w}_{K}\end{bmatrix}\mathbf{x}
= \begin{bmatrix}w_{11} &w_{12} & \cdots &w_{1D}\\
w_{21} &w_{22} &\cdots &w_{2D}\\
\vdots &\vdots &\ddots &\vdots\\
w_{K1} &w_{K2} &\cdots &w_{KD}
\end{bmatrix}
\begin{bmatrix} x_1 \\ x_2\\ \vdots\\ x_{D}\end{bmatrix}$$

The within-class covariance matrix in the projected $K$-dimensional $\mathbf{y}$ space is given by
$$s_{W} = \sum_{k=1}^K\sum_{n\in C_k}(\mathbf{y}_n-\mathbf{\mu}_k)(\mathbf{y}_n-\mathbf{\mu}_k)^T \tag{4.47}$$
The between-class covariance matrix in the projected $K$-dimensional $\mathbf{y}$ space is given by
$$s_{B} = \sum_{k=1}^K N_k(\mathbf{\mu}_k-\mathbf{\mu})(\mathbf{\mu}_k-\mathbf{\mu})^T \tag{4.48}$$
where
$$\mathbf{\mu}_{k} = \frac{1}{N_k}\sum_{n\in C_k}\mathbf{y}_n,\qquad \mathbf{\mu} = \frac{1}{N}\sum_{k=1}^K N_k\mathbf{\mu}_k \tag{4.49}$$

Again we wish to construct a scalar that is large when the between-class covariance is large and when the within-class covariance is small. There are now many possible choices of criterion (Fukunaga, 1990). One example is given by
$$J(\mathbf{W}) = Tr\{s_W^{-1}S_B\} \tag{4.50}$$
We need to maximize this creterion by adjusting $\mathbf{W}$.
