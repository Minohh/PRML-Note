A discriminant is a function that takes an input vector $\mathbf{x}$ and assigns i to one of $K$ classes, denoted $C_k$.

# Two classes

## Definition
The simplest linear discriminant function
$$y(\mathbf{x})=\mathbf{w}^T\mathbf{x}+w_0 \tag{4.4}$$
where
- $\mathbf{w}$ is called a *weight vector*.
- $w_0$ is a bias.
- $\mathbf{x}$ is an input vector.
- $y(\mathbf{x})$ is an output scalar value and is use to determine the class of the input $\mathbf{x}$. 
$$class\ of\ \mathbf{x}=\left\{\begin{array}{ll}C_1, &if\ y(\mathbf{x})\geqslant 0\\ C_2, &if\ y(\mathbf{x})\leqslant 0\end{array}\right .$$
So $y(\mathbf{x})=0$ is the decision boundary is a $(D-1)$-dimensional hyperplane with the $D$-dimensional input space.

$(4.4)$ can also be writen to
$$y(\mathbf{x})=\tilde{\mathbf{w}}^T\tilde{\mathbf{x}}\qquad where\ \left\{\begin{array}{ll}
\tilde{\mathbf{w}}=(w_0,\mathbf{w})\\
\tilde{\mathbf{x}}=(1,\mathbf{x})\\
\end{array}\right.\tag{4.8}$$


## Geometric description

Consider two points $\mathbf{x}_A$ and $\mathbf{x}_B$ lie on the decision surface.
$$y(\mathbf{x}_A) = y(\mathbf{x}_B) = 0\Rightarrow \mathbf{w}^T(\mathbf{x}_A-\mathbf{x}_B)=0$$
And hence
- $(\mathbf{x}_A-\mathbf{x}_B)$ represents the vector lying within the decision surface.
- $\mathbf{w}$ determines the orientation of the decision surface.
- The normal distance from the origin to the decision surface is given by
$$\frac{\mathbf{w}^T}{\|\mathbf{w}\|}\mathbf{x} = -\frac{w_0}{\|\mathbf{w}\|} \tag{4.5}$$
$w_0$ therefore determines the location of the decision surface.
- Consider an arbitrary vector $\mathbf{x}$. We analysis this vector $\mathbf{x}$ into two vectors
  - $\mathbf{x}_{\perp}$ is the orthogonal projection of point $\mathbf{x}$ onto the decision surface, which implicates that $\mathbf{x}_{\perp}$ is a point that lies on the decision surface.
  - $r\frac{\mathbf{w}}{\|\mathbf{w}\|}$ is the vector from the decision surface (point $\mathbf{x}_{\perp}$) to the point $\mathbf{x}$. It is parallel to the vector $\mathbf{w}$.
  - $r$ is the perpendicular distance from the the decision surface to the point $\mathbf{x}$.
$$\begin{align*}
\mathbf{x} &= \mathbf{x}_{\perp}+r\frac{\mathbf{w}}{\|\mathbf{w}\|}\\
\Rightarrow \mathbf{w}^T\mathbf{x}+w_0&= \mathbf{w}^T\mathbf{x}_{\perp}+r\frac{\mathbf{w}^T\mathbf{w}}{\|\mathbf{w}\|}+w_0\\
y(\mathbf{x}) &= \underbrace{y(\mathbf{x}_{\perp})}_{vanish}+r\|\mathbf{w}\|\\
\Rightarrow r &= \frac{y(\mathbf{x})}{\|\mathbf{w}\|} \tag{4.7}
\end{align*}$$
  - $y(\mathbf{x})$ therefore gives a signed measure of the perpendicular distance $r$ of the point $x$ from the decision surface.

```python
import numpy as np
import matplotlib.pyplot as plt

limX = [-2, 5]
limY = [-2, 5]

w0 = -4.0
w1 = 1.0
w2 = 1.5
x2 = lambda x1: -(w1*x1+w0)/w2

o = np.array([[0], [0]])
x = np.array([[4], [2]])

def main():
    fig = plt.figure(figsize=(6,6), dpi=60)
    ax = fig.add_subplot(1,1,1)
    
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    ax.set_xlabel(r"$x_1$", fontsize=14, x=1)
    ax.set_ylabel(r"$x_2$", fontsize=14, y=1)
    ax.set_xlim(limX[0], limX[1])
    ax.set_ylim(limY[0], limY[1])
    
    # Decision surface
    X = np.linspace(-1, 5, 10)
    Y = x2(X)
    ax.plot(X, Y, color='red')
    ax.text(x=-1.5, y=Y[0], s=r"$y = 0$", fontsize=14, color='red')
    ax.text(x=-1, y=Y[0]+0.5, s=r"$y > 0$", fontsize=14)
    ax.text(x=-2, y=Y[0]-0.5, s=r"$y < 0$", fontsize=14)
    ax.text(x=-1, y=Y[0]-0.7, s=r"$\mathcal{R}_1$", fontsize=15)
    ax.text(x=-0.5, y=Y[0]-0.2, s=r"$\mathcal{R}_2$", fontsize=15)
    
    # New point
    ax.quiver(o[0], o[1], x[0], x[1] , color='blue', angles='xy', scale_units='xy', scale=1)
    ax.text(x=x[0], y=x[1], s=r"$\mathbf{x}$", fontsize=14, color='blue')
    
    # Orientation vector
    w = np.array([[w1], [w2]])
    ax.quiver(o[0], o[1], w[0], w[1] , color='green', angles='xy', scale_units='xy', scale=1)
    ax.text(x=w[0], y=w[1], s=r"$\mathbf{w}$", fontsize=14, color='green')
    
    # Auxiliary lines
    px0 = limX[1]*0.5
    px1 = -w1*px0/w2
    ax.plot([0, px0], [0, px1], color='k')
    
    w_mod = np.sqrt(w[0]**2+w[1]**2)
    r = (w1*x[0][0]+w2*x[1][0]+w0)/w_mod
    x1x = x[0][0]-w[0]/w_mod*r
    x1y = x[1][0]-w[1]/w_mod*r
    ax.scatter(x1x, x1y, color='C1', s=50, marker='o')
    ax.text(x=x1x-0.6, y=x1y-0.1, s=r"$\mathbf{x}_{\perp}$", fontsize=14, color='C1')

    plt.annotate(
        '', xy=(x[0][0], x[1][0]), xycoords='data',
        xytext=(x1x, x1y), textcoords='data',
        arrowprops={'arrowstyle': '<->', 'color':'k'})
    ax.text(x=(x1x+x[0][0])/2+0.1, y=(x1y+x[1][0])/2-0.3, s=r"$r = \frac{y(\mathbf{x})}{\||\mathbf{w}\||}$", fontsize=14, color='k')

    r_orig = -w0/w_mod
    x2x = x1x-w[0]/w_mod*r_orig
    x2y = x1y-w[1]/w_mod*r_orig
    plt.annotate(
        '', xy=(x1x, x1y), xycoords='data',
        xytext=(x2x, x2y), textcoords='data',
        arrowprops={'arrowstyle': '<->', 'color':'k'})
    ax.text(x=(x1x+x2x)/2+0.1, y=(x1y+x2y)/2-0.3, s=r"$\frac{-w_0}{\||\mathbf{w}\||}$", fontsize=14, color='k')

if __name__ == "__main__":
    main()
```

# Multiple classes

We would briefly mention 3 approaches for classifing multiple classes here.
- *One-versus-the-rest* classifier. Use $K-1$ classifiers each of which solves a two-class problem of seperatin points in a particular class $C_k$ from points not in the class.
- *One-versus-one* classifier. Introduce $K(K-1)/2$ binary discriminant functions, one for every possible pair of classes.
- <font color='Red'>A single $K$-class discriminant comprising $K$ linear functions of the form
$$y_k(\mathbf{x}) = \mathbf{w}_k^T\mathbf{x}+w_{k0} \tag{4.9}$$
and then assigning a point $\mathbf{x}$ to class $C_k$ if $y_{k}(\mathbf{x})>y_j(\mathbf{x})$ for all $j\neq k$. The decision boundary between class $C_k$ and class $C_j$ is therefore given by $y_k(\mathbf{x})=y_j(\mathbf{x})$ and hencw corresponds to a $(D-1)$-dimensional hyperplane defined by 
$$(\mathbf{w}_k-\mathbf{w}_j)^T\mathbf{x}+(w_{k0}-w_{j0}) = 0 \tag{4.10}$$</font>

The first and second approaches always lead to regions of input space that are ambiguously classified, whereas the third approach is the approach that we need.
